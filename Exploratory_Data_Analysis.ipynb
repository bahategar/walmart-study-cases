```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./dataset/walmart.csv')
print(df.info())
df.head()
```

# General Customer and Product Sales

```python
summarize_user = df.groupby('User_ID')\
                    .agg({'Purchase': ['sum', 'count', 'mean', 'median', 'min', 'max']})\
                    .reset_index()\
                    .sort_values(by=[('Purchase','sum')], ascending=False)

summarize_user.head(25)
```

```python
cust_lifetime_value = df.groupby('User_ID')\
                        .agg({'Purchase': 'sum'})\
                        .reset_index()\
                        .sort_values(by='Purchase', ascending=False)


cust_freq = df.groupby('User_ID')\
                .size()\
                .reset_index()\
                .rename(columns={0: 'count'})\
                .sort_values(by='count', ascending=False)


top10_clv = cust_lifetime_value.head(10)
top10_cust_freq = cust_freq.head(10)

fig, axes = plt.subplots(1, 2, figsize=(20, 7))

axes[0].set_title("Top 10 Customer Lifetime Value")
sns.barplot(ax=axes[0], data=top10_clv,
            x='User_ID', y='Purchase',
            order=top10_clv.sort_values('Purchase', ascending=False).User_ID)
axes[0].set_ylabel("Customer Lifetime Value")
axes[0].set_xlabel("User ID")


axes[1].set_title("Top 10 Customer Frequency Order")
sns.barplot(ax=axes[1], data=top10_cust_freq,
            x='User_ID', y='count',
            order=top10_cust_freq.sort_values('count', ascending=False).User_ID)
axes[1].set_ylabel("Frequency Order")
axes[1].set_xlabel("User ID")

plt.show()
```

```python
summarize_product = df.groupby('Product_ID')\
                    .agg({'Purchase': ['sum', 'count', 'mean', 'median', 'min', 'max']})\
                    .reset_index()\
                    .sort_values(by=[('Purchase','sum')], ascending=False)

summarize_product.head(25)
```

```python
product_sales = df.groupby('Product_ID')\
                        .agg({'Purchase': 'sum'})\
                        .reset_index()\
                        .sort_values(by='Purchase', ascending=False)


product_freq = df.groupby('Product_ID')\
                .size()\
                .reset_index()\
                .rename(columns={0: 'count'})\
                .sort_values(by='count', ascending=False)


top10_product_sales = product_sales.head(10)
top10_product_freq = product_freq.head(10)

fig, axes = plt.subplots(1, 2, figsize=(24, 7))

axes[0].set_title("Top 10 Purchasing Products")
sns.barplot(ax=axes[0], data=top10_product_sales,
            x='Product_ID', y='Purchase',
            order=top10_product_sales.sort_values('Purchase', ascending=False).Product_ID)
axes[0].set_ylabel("Customer Lifetime Value")
axes[0].set_xlabel("Product ID")


axes[1].set_title("Top 10 Frequency Order Products")
sns.barplot(ax=axes[1], data=top10_product_freq,
            x='Product_ID', y='count',
            order=top10_product_freq.sort_values('count', ascending=False).Product_ID)
axes[1].set_ylabel("Frequency Order")
axes[1].set_xlabel("Product ID")

plt.show()
```

# Urban and Rural Analysis

```python
# Create a figure with two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(20, 7))

# First plot: Total Purchasing by City (and long stay)
axes[0].set_title("Total Purchasing by City (and long stay)")
hue_order = sorted(df['Stay_In_Current_City_Years'].unique())
sns.barplot(ax=axes[0], data=df, y='City_Category', x='Purchase',
            estimator='sum', hue='Stay_In_Current_City_Years',
            hue_order=hue_order)
axes[0].set_ylabel('City')
axes[0].set_xlabel('Total Purchasing')
axes[0].legend(loc='upper right')

# Second plot: Total Order by City (and long stay)
axes[1].set_title("Total Order by City (and long stay)")
sns.countplot(ax=axes[1], data=df, y='City_Category',
              hue='Stay_In_Current_City_Years',
              hue_order=hue_order)
axes[1].set_ylabel('City')
axes[1].set_xlabel('Total Orders')
axes[1].legend(loc='upper right')


# Show the combined plots
plt.show()
```

```python
count_data = df.groupby(['Product_Category', 'City_Category']).size().unstack(fill_value=0)
percentage_data = round(count_data.div(count_data.sum(axis=1), axis=0) * 100, 3)

fig, axes = plt.subplots(1, 2, figsize=(20, 7))

count_data.plot(kind='bar', stacked=True, ax=axes[0])
axes[0].set_title("Distribution Product Category per City (Count)")
axes[0].set_xlabel("Product Category")
axes[0].set_ylabel("Frequency")
axes[0].legend(title='City Category', loc='upper right')

  
percentage_data.plot(kind='bar', stacked=True, ax=axes[1])
axes[1].set_title("Distribution Product Category per City (Percentage)")
axes[1].set_xlabel("Product Category")
axes[1].set_ylabel("Percentage Frequency")
axes[1].legend(title='City Category', loc='upper right')


plt.show()
```

```python
count_data = df.groupby(['Occupation', 'City_Category']).size().unstack(fill_value=0)
percentage_data = round(count_data.div(count_data.sum(axis=1), axis=0) * 100, 3)

fig, axes = plt.subplots(1, 2, figsize=(20, 7))

count_data.plot(kind='bar', stacked=True, ax=axes[0])
axes[0].set_title("Distribution Occupation per City (Count)")
axes[0].set_xlabel("Occupation")
axes[0].set_ylabel("Frequency")
axes[0].legend(title='City Category', loc='upper right')


percentage_data.plot(kind='bar', stacked=True, ax=axes[1])
axes[1].set_title("Distribution Occupation per City (Percentage)")
axes[1].set_xlabel("Occupation")
axes[1].set_ylabel("Percentage Frequency")
axes[1].legend(title='City Category', loc='upper right')


plt.show()
```

```python
count_data = df.groupby(['Age', 'City_Category']).size().unstack(fill_value=0)
percentage_data = round(count_data.div(count_data.sum(axis=1), axis=0) * 100, 3)

fig, axes = plt.subplots(1, 2, figsize=(20, 7))

count_data.plot(kind='bar', stacked=True, ax=axes[0])
axes[0].set_title("Distribution Age per City(Count)")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Frequency")
axes[0].legend(title='City Category', loc='upper right')


percentage_data.plot(kind='bar', stacked=True, ax=axes[1])
axes[1].set_title("Distribution Age per City (Percentage)")
axes[1].set_xlabel("Age")
axes[1].set_ylabel("Percentage Frequency")
axes[1].legend(title='City Category', loc='upper right')


plt.show()
```

```python
# Create a figure with two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(20, 7))

marital_purchase = df.groupby("Marital_Status").agg({"Purchase": "sum"}).reset_index()
axes[0].set_title("Total Purchasing by Marital Status")
sns.barplot(ax=axes[0], data=marital_purchase,
            x='Marital_Status', y='Purchase',)
axes[0].set_ylabel('Marital Status')
axes[0].set_xlabel('Total Purchasing')

axes[1].set_title("Total Order by Marital Status")
sns.countplot(ax=axes[1], data=df, x='Marital_Status',
              hue='City_Category', hue_order=sorted(df['City_Category'].unique()))
axes[1].set_ylabel('Marital Status')
axes[1].set_xlabel('Total Orders')


# Show the combined plots
plt.show()
```

# Occupational Influence on Purchases

```python
# Create a figure with two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(20, 7))

occupation_purchase = df.groupby("Occupation").agg({"Purchase": "sum"}).reset_index()
axes[0].set_title("Total Purchasing by Occupation")
sns.barplot(ax=axes[0], data=occupation_purchase,
            x='Occupation', y='Purchase',
            palette=sns.color_palette("tab10"))
axes[0].set_ylabel('Occupation')
axes[0].set_xlabel('Total Purchasing')

axes[1].set_title("Total Order by Occupation")
sns.countplot(ax=axes[1], data=df, x='Occupation', palette=sns.color_palette("tab10"))
axes[1].set_ylabel('Occupation')
axes[1].set_xlabel('Total Orders')


# Show the combined plots
plt.show()
```
